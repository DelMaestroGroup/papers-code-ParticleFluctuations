```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pair_c = np.loadtxt('../data/pair-T-reduce-243e1e6f-6003-49ae-8d1f-3204119c6e64.dat',skiprows=2)
pair_gc = np.loadtxt('../data/pair-T-reduce-38860f44-184f-4317-ba4f-787b35130dc8d.dat',skiprows=2)
```

```python
with plt.style.context('../include/aps.mplstyle'):
    figsize = plt.rcParams['figure.figsize']
    fig, ax = plt.subplots(constrained_layout=True, figsize=(figsize[0],figsize[1]))
    ax.errorbar(pair_gc[:,0]/100,pair_gc[:,1],yerr = pair_gc[:,2], color='blue',label=r'Grand canonical $\langle N\rangle = 25.01$')
    ax.errorbar(pair_c[:,0]/100,pair_c[:,1],yerr = pair_c[:,2],color='orange',label=r'canonical, $N = 25$')
    ax.set_xlabel(r'$x/L$')
    ax.set_ylabel(r'$\langle \rho(x)\rho(0)\rangle/\rho_0^2$')
    ax.legend(loc = 1)
    ax.set_xlim(0,0.49)
    ax.grid(True, alpha=0.3)
    plt.savefig('../figures/1DHePairCorr.pdf')
```
